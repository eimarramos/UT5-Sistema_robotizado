```python
from ultralytics import YOLO
import os

# Ruta absoluta al yaml
yaml_path = os.path.abspath("dataset/data.yaml")

model = YOLO("yolo11n.pt")

model.train(
    data=yaml_path,
    epochs=50,
    imgsz=416,
    batch=16,
    name="chess_detector"
)
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/chess_detector/weights/best.pt")

metrics = model.val()
print(f"mAP50:    {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")
```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/chess_detector/weights/best.pt")

CLASES = [
    'bishop', 'black-bishop', 'black-king', 'black-knight',
    'black-pawn', 'black-queen', 'black-rook', 'white-bishop',
    'white-king', 'white-knight', 'white-pawn', 'white-queen', 'white-rook'
]

results = model.predict(
    source="dataset/test/images",
    imgsz=416,
    conf=0.25,
    save=True,
    show=True
)

# Mostrar detecciones por imagen
for r in results:
    print(f"\nImagen: {r.path}")
    for box in r.boxes:
        cls_id = int(box.cls)
        conf   = float(box.conf)
        print(f"  → {CLASES[cls_id]}: {conf:.2%} confianza")
```
